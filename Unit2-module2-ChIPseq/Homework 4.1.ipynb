3. For all the alignment output sam files:

3.1 convert to bam, sort and filter

```python
%%bash
cd /home/rtu/Sample_fastq_data/Sample_fastq_trimmed

for i in *.sam #test for loop. it works.
do
samtools view -bS $i > ${i:0:10}.bam #$i from indexes 0:10 is the SRR file name
samtools sort ${i:0:10}.bam -o ${i:0:10}.srt.bam
samtools rmdup -S ${i:0:10}.srt.bam ${i:0:10}.srt.dupr.bam

done
```
