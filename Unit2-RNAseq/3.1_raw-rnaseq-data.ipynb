Run first 5 SRR runs from https://www.ncbi.nlm.nih.gov/Traces/study/?acc=SRP112343

````python
%%bash
cd /home/rtu/
for((a=5830636; a<=5830640 ;a++))
do
    ./sratoolkit.2.9.2-ubuntu64/bin/fastq-dump --split-files -X 100 -Z SRR$a > /home/rtu/Sample_fastq/SRR$a.100.fastq
done

#First 5 Runs
````

Compare the ATCG's from all five outputs.

````python
%%bash
cd /home/rtu/Sample_fastq/
#A's
echo "# of A's for runs 1 - 5"
for ((a=5830636;a<=5830640;a++))
    do
    awk 'NR%4==2' SRR$a.100.fastq | tr -cd 'A' |wc -c
done

#T's
echo "# of T's for runs 1 - 5"
for ((a=5830636;a<=5830640;a++))
    do
    awk 'NR%4==2' SRR$a.100.fastq | tr -cd 'T' |wc -c
done

#C's
echo "# of C's for runs 1 - 5"
for ((a=5830636;a<=5830640;a++))
    do
    awk 'NR%4==2' SRR$a.100.fastq | tr -cd 'C' |wc -c
done

#G's
echo "# of G's for runs 1 - 5"
for ((a=5830636;a<=5830640;a++))
    do
    awk 'NR%4==2' SRR$a.100.fastq | tr -cd 'G' |wc -c
done
````

All runs have 15,200 base pairs. "A" seems to be more frequent in each run than other nucleotides. One would expect A's to match with the amount of T's and the C's to correspond to the same number of G's in DNA. However, since this is RNA, you don't need to have a complete match between your corresponding nucleotides. Taking all the AT's and GC's in every read, the AT's correspond to 53, 50, 49, 52, and 51% of the nucleotide composition, which is not totally abnormal in count. Potential reasons for mismatched reads are different quality scores of the reads which can skew distribution.

When the read aligns to chromosome 'star', there isn't any alignment of that particular read to the genome.  When the read aligns to chromosome MT, it corresponts to alignment to mitochondrial DNA. When the output is a KI###, it corresponds to a loci in a gene.
